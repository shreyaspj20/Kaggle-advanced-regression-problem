# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
df=train.append(test)
```

```python
num=df.select_dtypes(include=np.number)
```

```python
cat=df.select_dtypes(exclude=np.number)
```

# Imputing values for categorical and numerical variables

```python
num['LotFrontage'].fillna(68,inplace=True)
num['GarageYrBlt'].fillna(1979,inplace=True)
num.fillna(0,inplace=True)
print(num.isnull().sum())
```

```python
cat.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp =SimpleImputer(missing_values=np.nan, strategy='most_frequent').fit(cat)
cat1=imp.transform(cat)
cat
```

```python
cat2=pd.DataFrame(cat1)
cat2
```

```python
cat2.columns=cat.columns
```

```python
cat2.isnull().sum()
cat2=pd.get_dummies(cat2)
```

```python
import xgboost
from xgboost import plot_importance
model = xgboost.XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.07,
                 max_depth=3,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42) 
```

```python
print(cat2.shape)
print(num.shape)
```

```python
d=num.reset_index(drop=True).merge(cat2.reset_index(drop=True), left_index=True, right_index=True)
```

```python
train_dataset =d.iloc[:1460,:]
test_dataset =d.iloc[1460:,:]
```

# Feature Engineering

```python
every_column_except_y= [col for col in train_dataset.columns if col not in ['SalePrice','Id']]
model.fit(train_dataset[every_column_except_y],train_dataset['SalePrice'])
```

```python
most_relevant_features= list( dict((k, v) for k, v in model.get_booster().get_fscore().items() if v >= 20).keys())
print(most_relevant_features)
```

# Hyperparameter tuning using GridSearchCV

```python

from sklearn.model_selection import GridSearchCV 
x_train=train_dataset[most_relevant_features]
y_train=train_dataset['SalePrice']

best_xgb_model = xgboost.XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.07,
                 max_depth=3,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42)
best_xgb_model.fit(x_train,y_train)
```

```python
x_test=test_dataset[most_relevant_features]
y_pred=best_xgb_model.predict(x_test)
```

```python
submission=pd.DataFrame({'Id':test_dataset['Id'],'SalePrice':y_pred})
```
